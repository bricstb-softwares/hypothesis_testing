```python
#Importing image data for geodesic evaluation
def get_transform(params=None, grayscale=False, method=Image.BICUBIC, convert = True):
    transform_list = []
    if grayscale:
        transform_list.append(transforms.Grayscale(1))
    if 'resize' in params:
        osize = [load_size, load_size]
        transform_list.append(transforms.Resize(osize, method))
    elif 'scale_width' in params:
        transform_list.append(transforms.Lambda(lambda img: __scale_width(img, load_size, crop_size, method)))
    if convert:
        transform_list += [transforms.ToTensor()]
        if grayscale:
            transform_list += [transforms.Normalize((0.5,), (0.5,))]
        else:
            transform_list += [transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))]
    return transforms.Compose(transform_list)

def __scale_width(img, target_size, crop_size, method=Image.BICUBIC):
    ow, oh = img.size
    if ow == target_size and oh >= crop_size:
        return img
    w = target_size
    h = int(max(target_size * oh / ow, crop_size))
    return img.resize((w, h), method)
```

```python
#Class for load and access the data
class BaseDataset(data.Dataset):
    """Face Landmarks dataset."""

    def __init__(self, fileguide, transform=None):
        """
        Arguments:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        #TO-DO create a task for reading by filepath
        #if root_csv_file is not None: 
        #    self.fileguide = pd.read_csv(root_csv_file)
        self.fileguide = fileguide['image_path'].tolist()
        self.fileguide_size = len(self.fileguide)
        self.transform = transform

    def __len__(self):
        return len(self.fileguide)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_name = self.fileguide[idx % self.fileguide_size]
        image = Image.open(img_name).convert('RGB')
        image = ImageOps.exif_transpose(image)
        if self.transform:
            im = self.transform(image)
        #sample = {'image': im, 'path': img_name}
        return im
```

# Statiscal Functions

```python
def calc_chi_s(counts_obs_hist, counts_ref_hist):
    '''
    This function will calculate the $\chi^s$ statistic for a given two histograms

    Arguments:

    - counts_obs_hist: coints of observed numpy histogram.
    - counts_ref_hist: counts of reference numpy histogram.
    '''

    return (counts_obs_hist - counts_ref_hist)/np.sqrt(counts_ref_hist)

def calc_chi_s_residuals(counts_obs_hist, counts_ref_hist):
    '''
    This function will calculate the $\chi^s$ residuals for $\chi^s$ statistic.

    Arguments:

    - counts_obs_hist: coints of observed numpy histogram.
    - counts_ref_hist: counts of reference numpy histogram.

    '''
    diff_counts = (counts_obs_hist.sum() - counts_ref_hist.sum())
    return (diff_counts*(counts_ref_hist/counts_ref_hist.sum()))/np.sqrt(counts_ref_hist)

def calc_kl(pk, qk):
    '''
    A function to calculate the Kullback-Libler divergence between p and q distribution.
    Arguments:
    pk: pdf values from p distribution
    qk: pdf values from q distribution
    '''
    return scipy.special.kl_div(pk, qk)

def calc_js(pk, qk):
    '''
    A function to calculate the Jensen-Shanon divergence between p and q distribution.
    Arguments:
    pk: pdf values from p distribution
    qk: pdf values from q distribution
    '''
    mk = 0.5*(pk+qk)
    return 0.5*(calc_kl(pk, mk) + calc_kl(qk, mk))

def calc_wasserstein(pk, qk):
    
    cdf_p = np.cumsum(pk)
    cdf_q = np.cumsum(qk)
    return np.abs(cdf_p - cdf_q)
```

# Non-Parametric Approach

```python
class non_parametric_conection:
    def __init__(self, step, pdf1, pdf2, support):
        '''
            This implementation will create connections (e and m) between two PDFs estimated from a Gaussian KDE with a adjustable bandwidth.
            Arguments:
                - step: number of t paramters obtained from the connections.
                - pdf1: the pdf used to be the intial point for build the connections.
                - pdf2: the pdf used to be the final point for build the connections.
                - support: the binning i.e. the values that should be used to integrate
                NOTE: The pdf1 and pdf2 must have the same support.

        '''
        
        self.step = step
        #self.lim_inf = lim_inf
        #self.lim_sup = lim_sup
        self.p1 = pdf1
        self.p2 = pdf2
        self.support = support
        self.dx = np.diff(self.support)[:-1]
        self.t_params = np.arange(0, 1+self.step, self.step)
        self.exp_connection = self.get_exp_connection()
        
    
    def get_exp_connection(self):
        '''
        This function will compute the exponential connection as described in the Amari's Information Geometry and Its Applications book.
        '''
        self.e_geo = []
        for it in self.t_params:
            con   = (1-it)*np.log(self.p1) + it*np.log(self.p2)
            phi_t = np.log(np.trapz(np.exp(con), dx=self.dx))
            self.e_geo.append(np.exp(con - phi_t))
        return self.e_geo
    
    def phi_t(self, t):
        con   = (1-t)*np.log(self.p1) + t*np.log(self.p2)
        phi_t = np.log(np.trapz(np.exp(con), dx=self.dx))
        
    def get_mix_connection(self):
        '''
        This function will compute the exponential mixture as described in the Amari's Information Geometry and Its Applications book.
        '''
        self.mix = []
        for it in self.t_params:
            self.mix.append((1-it)*self.p1 + it*self.p2)
        return self.mix
    
    def get_tangent_vectors(self, data_p3, p3_index_projection):
        '''
        This function will compute the tangent vectors for the e-connection and m-connection as described in the Amari's Information Geometry and Its Applications book given a 3th point.
        Argument:
            - data_p3: a new data that could be between the intial and final points of connections.
            - p3_index_projection: the index used to be a projection of the data_p3 in the connections.
        '''
        # estimate from p3 data
        if self.pdf_estimation_method == 'kde':
            data_p3 = data_p3[(data_p3>=self.lim_inf) & (data_p3<self.lim_sup)]
            pdf_model3 = scipy.stats.gaussian_kde(data_p3, bw_method=self.bandwidth)
            pdf_p3 = pdf_model3.pdf(self.support)
        elif self.pdf_estimation_method == 'hist':
            p3_count, p3_supp = np.histogram(data_p3, bins=self.binning, density=True)
            pdf_k     = p3_count
            
        # calculate the tangent vectors
        return np.log(self.p2)-np.log(self.p1), self.e_geo[p3_index_projection]-pdf_p3
    
    def get_nearest_point(self, pdf3):
        '''
        This function will compute the nearest point to the connection given a 3rd PDF in terms of Kullback-Leibler divergence.
        Argument:
            - pdf3: a new pdf that remains in the same manifold but probably isn't in the connection.
        '''

        best_match = 100
        best_idx   = -1
        con3   = np.log(pdf3)
        phi_t3 = np.log(np.trapz(np.exp(con3), dx=self.dx))
        pdf3   = np.exp(con3 - phi_t3)
        
        all_kl = []
        for idx in range(len(self.exp_connection)):
            kl_values = calc_kl(self.exp_connection[idx], pdf3)
            total_kl  = np.trapz(kl_values, dx=self.dx)
            all_kl.append(total_kl)
            if total_kl < best_match: 
                best_match = total_kl
                best_idx   = idx
        return best_match, best_idx, np.array(all_kl)
```

```python
t_min_obs_i = {}
t_bound_obs_i = {}
for i in range(len(histograms)):
    if i == 10:
        continue
    t_min_obs_j = []
    t_bound_obs_j = []
    for j in range(len(histograms[0])):
        manifold = non_parametric_conection(step=1e-3,
                                                      pdf1=histograms[i][j][0],
                                                      pdf2=histograms[i][j][2],
                                                      support=histograms[i][j][3])
        
        #Get projections values
        
        t_best_kl, t_best_idx, t_kl_values = manifold.get_nearest_point(histograms[i][j][0])
        
        v_best_kl, v_best_idx, v_kl_values = manifold.get_nearest_point(histograms[i][j][2])
        
        m_best_kl, m_best_idx, m_kl_values = manifold.get_nearest_point(histograms[i][j][1])

        t_min_obs_j.append(m_best_idx)
        
        # Get boundaries
        
        df_spectrum = pd.DataFrame.from_dict({'params': manifold.t_params, 't_values':  t_kl_values, 'v_values':  v_kl_values})
        
        df_spectrum['diff_values'] = abs(df_spectrum['v_values'] - df_spectrum['t_values'])
        
        if df_spectrum.loc[ df_spectrum['diff_values'] < 0.0001 , 'params'].empty:
            
            t_bound = np.median(df_spectrum.loc[ df_spectrum['diff_values'] < 0.001 , 'params'].values)
        else:
            t_bound = df_spectrum.loc[ df_spectrum['diff_values'] < 0.0001 , 'params'].values[0] * 1000

        t_bound_obs_j.append(t_bound)
        
    t_min_obs_i[i] = t_min_obs_j
    t_bound_obs_i[i] = t_bound_obs_j
t_min_df = pd.DataFrame.from_dict(t_min_obs_i)
t_bound_df = pd.DataFrame.from_dict(t_bound_obs_i)
```
